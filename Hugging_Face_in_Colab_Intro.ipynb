```python
from transformers import pipeline
```

# Initialise a pipeline
### Here we initialise a sentiment analysis pipeline


```python
classifier = pipeline("sentiment-analysis")
result = classifier("I love using Hugging Face models in Colab!")
print(result)
```

```python
# Selecting a particular model for a pipeline
classifier = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")
```

# Other tasks and models

```python
# Text Generation
generator = pipeline("text-generation", model="gpt2")
result = generator("Once upon a time,")
print(result)
```

```python
# Translation
translator = pipeline("translation_en_to_fr")
result = translator("I love coding in Python!")
print(result)
```

```python
# Question Answering
question_answerer = pipeline("question-answering")
result = question_answerer({
      "question": "What is the capital of France?",
      "context": "Paris is the capital of France."
})
print(result)
```
